# ARIMA Time Series Forecasting Preprocessing and EDA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

## EDA and Preprocessing

```python
# Load the data
data=pd.read_csv('New motor vehicle sales data.csv')

# Display the shape of the data
data.shape
```

```python
# Display the first few rows of the data
data.head()
```

### Checked missing value

since no missing value in the main columns we need "REF_DATE" and "Value", no need to drop any missing value

```python
# check for missing values
data.isnull().sum()
```

### Check duplicates

```python
# Check for duplicates
data.duplicated().sum()
```

### Deal with aggregated data

We found that in the dataset, there is aggregated 
1. GEO: national data and provincial data.
2. Vehicle type: Passenger, Truck, and Total New
3. Origin of manufacture: "Japen", "North America", "Other Countries", "Total, country of manufacture", "Total, overseas"


```python
# Display the unique value within each columns
data.nunique()
```

#### 1.Deal with aggregated GEO

```python
# Display the  unique values in the 'GEO' column
data['GEO'].unique()
```

```python
# Split the data with 'GEO="Canada"'
data_canada=data[data['GEO']=='Canada']

# Display the shape of the data
data_canada.head()
```

```python
# Display the shape of the data
data_canada.shape
```

```python
#Split for provincial data 'GEO!="Canada"
data_provincial=data[data['GEO']!='Canada']

#Display the unique values in the 'GEO' column to confirm that the rows with 'GEO="Canada" have been removed
# data['GEO'].unique()

#Display the shape of the data
data_provincial.shape
```

```python
# Display the first few rows of the data
data_provincial.head()
```

#### 2.Deal with aggregated Vehicle type

```python
# Display the  unique values in the 'Vehicle type' column
data_canada['Vehicle type'].unique()
```

```python
data_canada.shape
```

```python
# create a new df that only contains 'Total'
data_canada_vehicleType=data_canada[data_canada['Vehicle type']=='Total, new motor vehicles']
data_canada_vehicleType.shape
```

```python
# Display the unique values in the 'Vehicle type' column
data_provincial['Vehicle type'].unique()
```

```python
data_provincial.shape
```

```python
# create a new df that only contains 'Passenger cars','Trucks'
data_provincial_vehicleType=data_provincial[data_provincial['Vehicle type']=='Total, new motor vehicles']
data_provincial_vehicleType.shape
```

#### 3.Deal with aggregated Country of Origin

```python
data_canada_vehicleType['Origin of manufacture'].unique()
```

```python
data_canada_vehicleType_country = data_canada_vehicleType[data_canada_vehicleType['Origin of manufacture']=='Total, country of manufacture']
data_canada_vehicleType_country.shape
```

```python
# create a new df that only contains Japan, North America, Other Countries
data_provincial_vehicleType_country = data_provincial_vehicleType[data_provincial_vehicleType['Origin of manufacture']=='Total, country of manufacture']
data_provincial_vehicleType_country.shape
```

```python
# filter the provincial and canadian data for UOM = 'units'
data_canada_vehicleType_country_units = data_canada_vehicleType_country[data_canada_vehicleType_country['UOM']=='Units']
data_provincial_vehicleType_country_units = data_provincial_vehicleType_country[data_provincial_vehicleType_country['UOM']=='Units']
```

### Plot

```python
# Plot the "REF_DATE" and "Value" pair for each "GEO"

# Adjust figure size
plt.figure(figsize=(20, 10))

# Plot the lineplot
sns.lineplot(data=data_provincial_vehicleType_country_units, x='REF_DATE', y='VALUE', hue='GEO')

# Add more space for y-axis labels
plt.ylim(bottom=0)

# Set labels and title
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales vs Year for each Provinces')

# Show the plot
plt.show()
```

```python
# Adjust figure size
plt.figure(figsize=(25, 6))

# Format the date column to display only the last 2 digits of the year
data_canada['REF_DATE'] = pd.to_datetime(data_canada['REF_DATE']).dt.strftime('%y')

# Plot the lineplot
sns.lineplot(data=data_canada_vehicleType_country_units, x='REF_DATE', y='VALUE')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Value vs Year for Canada')
plt.show()
```

# Export processed file

```python
data_canada_vehicleType_country.to_csv('Canadian Sales.csv', index=False)
data_provincial_vehicleType_country.to_csv('Provincial Sales.csv', index=False)
```
